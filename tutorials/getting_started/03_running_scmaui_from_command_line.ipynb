This example shows how to run scMaui from the command line.

```python
import os
import pkg_resources
```

```python
data_path = pkg_resources.resource_filename('scmaui', 'resources/')
gtx = os.path.join(data_path, 'gtx.h5ad')
peaks = os.path.join(data_path, 'peaks.h5ad')
#files = [gtx,peaks]
print([gtx, peaks])
```

```python
!scmaui \
 -data {gtx} {peaks} \
 -datanames gtx peaks \
 -losses negbinom negmul \
 -output ./test_output
```
